## Ejercicio 1
Escribe una función que convierta números del 1 al 7 en nombres de los dias de la semana. La función constará de un único argumento numérico y una salida de tipo string

````python
def convertir_semana(num):
    dias_semana = ["Lunes", "Martes", "Miercoles", "Jueves", "Viernes", "Sabado", "Domingo"]
    for i, dia in enumerate(dias_semana):
        if i == (num-1):
            return dia

numero = 4
print(f"Dia de la semana para num {numero}: ", convertir_semana(numero))
````

## Ejercicio 3
Escribe una función que compare dos números. La función tiene dos argumentos y hay tres salidas posibles: que sean iguales, que el primero sea  mayor que el segundo, o que el segundo sea mayor que el primero

````python
def comparar_numeros(a, b):
    if a == b:
        return "Son iguales"
    elif a > b:
        return "El primero es mayor que el segundo"
    elif a < b:
        return "El segundo es mayor que el primero"
    else:
        return "?????"

print(comparar_numeros(1, 1))
print(comparar_numeros(4, 5))
print(comparar_numeros(10, 5))
print(comparar_numeros("a", "b"))
````

## Ejercicio 4
Escribe una función que sea un contador de letras. En el primer argumento tienes que introducir un texto, y el segundo que sea una letra que contar. La función tiene que devolver un entero con el número de veces que aparece esa letra, tanto mayúscula, como minúscula

````python
def contar_letras(texto, letra):
    texto_min = texto.lower()
    letra_min = letra.lower()

    texto_lista = list(texto_min)
    n_letras = texto_lista.count(letra_min)

    return n_letras

frase_random = "Hola que tal? COMO ESTAS?"
print("Frase elegida: ", frase_random)
print("Numero de letras 'a': ", contar_letras(frase_random, "a"))
print("Numero de letras 'e': ",contar_letras(frase_random, "e"))
````

## Ejercicio 5
Escribe una función que tenga un único argumento, un string. La salida de la función tiene que ser un diccionario con el conteo de todas las letras de ese string, siendo la clave la letra y el valor el conteo.

````python
def contar_aun_mas_letras(texto):
    texto_min = texto.lower()
    texto_lista = list(texto_min)
    counts_dict = {}
    letras_del_castellano = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 
                             'ñ', 'o', 'p', 'q', 'r', 's','t', 'u', 'v', 'w', 'x', 'y', 'z']
    for i in texto_lista:
        if i in letras_del_castellano:        
            if i in counts_dict:
                counts_dict[i] += 1
            else:
                counts_dict[i] = 1
    
    return counts_dict

counts_dict = contar_aun_mas_letras("Hola que tal? COMO ESTAS?")
print(counts_dict)
````

## Ejercicio 6
Escribir una función que añada o elimine elementos en una lista. La función necesita los siguientes argumentos:
* lista: la lista donde se añadirán o eliminarán los elementos
* comando: "add" o "remove"
* elemento: Por defecto es None.

Tendrá como salida la lista

````python
def anadir_o_eliminar(lista, comando, elemento=None):
    if comando == "add":
        lista.append(elemento)
    elif comando == "remove":
        lista.remove(elemento)
    else:
        print("Introduce un comando valido de los siguientes: 'add' o 'remove'")

    return None

lista6 = ["elemento aleatorio"]
anadir_o_eliminar(lista6, "add", 1)
print(lista6)
anadir_o_eliminar(lista6, "add", ["3", 5])
print(lista6)
anadir_o_eliminar(lista6, "remove", "elemento aleatorio")
print(lista6)
anadir_o_eliminar(lista6, "add")
print(lista6)
````

## Ejercicio 7
Crea una función que reciba un número arbitrario de palabras, y devuelva una frase completa, separando las palabras con espacios.

````python
def construir_frase(*palabras):
    frase = None
    for p in palabras:
        if frase != None:
            frase = frase + " " + p
        else:
            frase = p

    return frase

print("Frase construida: ", construir_frase("Hola", "que", "tal?", "COMO", "ESTAS?"))
````

## Ejercicio 8 - TEMARIO SIN VER EN CLASE!!!!
Escribe un programa que obtenga el enésimo número de la [serie de Fibonacci](https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci). Tienes que crear una función recursiva con un único argumento.

```Python
fibonacci(2) = 1
fibonacci(4) = 3
fibonacci(5) = 5
fibonacci(8) = 21
```

fibonacci(n) = fibonacci(n-1) + fibonacci(n-2)

````python
def fibotruco(n):
    if n < 1:
        return 0
    elif n == 1:
        return 1
    else:
        return fibotruco(n-1) + fibotruco(n-2)

print("Tercer elemento de la serie de Fibonacci: ", fibotruco(3))
print("Noveno elemento de la serie de Fibonacci: ", fibotruco(9))
````

* Un cuadrado de lado = 10 + 3 círculos (uno de radio = 4 y los otros dos de radio = 6) + 5 triángulos de base = 2 + altura = 4

````python
import math 
def area_cuadrado(a):
    return a ** 2

def area_triangulo(a, b):
    return (a * b)/2

def area_circulo(r):
    return math.pi*(r ** 2)
````

````python
area_total_1 = (2* area_circulo(10)) + area_triangulo(3, 7)
print("Area_total 1: ", round(area_total_1, 3))
area_total_2 = area_cuadrado(10) + area_circulo(4) + (2*area_circulo(6)) + (5*area_triangulo(2, 4))
print("Area_total 2: ", round(area_total_2, 3))
````
