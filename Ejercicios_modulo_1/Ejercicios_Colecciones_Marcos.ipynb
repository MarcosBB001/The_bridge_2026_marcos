## Ejercicio 1
Dada la siguiente lista:
> ```ejer_1 = [1,2,3,4,5]```

Inviértela para que quede de la siguiente manera
> ```ejer_1 = [5,4,3,2,1]```

````python
ejer_1 = [1,2,3,4,5]
ejer_1_reversed = ejer_1.copy()
for i, e in enumerate(ejer_1):
    indice = -i - 1
    ejer_1_reversed[indice] = e


print(ejer_1_reversed)
````

## Ejercicio 3
Crea una lista nueva con todas las combinaciones de las siguientes dos listas:
> ```ejer_3_1 = ["Hola", "amigo"]```
>
> ```ejer_3_2 = ["Que", "tal"]```

Obten el siguiente output:

```['Hola Que', 'Hola tal', 'amigo Que', 'amigo tal']```

````python
ejer_3_1 = ["Hola", "amigo"]
ejer_3_2 = ["Que", "tal"]

combinaciones = []
for palabra in ejer_3_1:
    for palabra2 in ejer_3_2:
        frase = palabra + " " + palabra2
        combinaciones.append(frase)

print(combinaciones)
````

## Ejercicio 4
Dada la siguiente lista, encuentra por el índice el valor 45 y sustitúyelo por el 0
> ```ejer_4 = [20, 47, 19, 29, 45, 67, 78, 90]```

## Ejercicio 5
Dada la siguiente lista, elimina todos los valores iguales a 3
> ```ejer_5 = [3, 20, 3, 47, 19, 3, 29, 45, 67, 78, 90, 3, 3]```

TIP: No intentes eliminar elementos sobre la lista que estás iterando. Haz una copia con `ejer_5.copy()`.

````python
ejer_5 = [3, 20, 3, 47, 19, 3, 29, 45, 67, 78, 90, 3, 3]
ejer_5_copia = ejer_5.copy()

for num in ejer_5:
    if num == 3:
        ejer_5_copia.remove(num)

print(ejer_5_copia)
    

````

## Ejercicio 11
Convierte la siguiente lista en un diccionario
> ```ejer_11 = [("x", 1), ("x", 2), ("x", 3), ("y", 1), ("y", 2), ("z", 1)]```

TIP: Puedes usar dos variables de control en un bucle `for`. Una por cada valor de las tuplas.

````python
ejer_11 = [("x", 1), ("x", 2), ("x", 3), ("y", 1), ("y", 2), ("z", 1)]
my_dict = {}
for e in ejer_11:
    if e[0] in my_dict:
        my_dict[e[0]].append(e[1])
    else:
        my_dict[e[0]] = [e[1]]
    

print(my_dict)
````

## Ejercicio 12
1. Crea una lista ordenada ascendente con las claves del diccionario
2. Crea otra lista ordenada descendente con los valores
3. Añade una nueva clave/valor
4. Busca la clave igual a 2 dentro del diccionario
5. Itera la clave y el valor del diccionario con un único for
> ```ejer_12 = {4:78, 2:98, 8:234, 5:29}```

````python
ejer_12 = {4:78, 2:98, 8:234, 5:29}

key_list = []
for k in ejer_12.keys():
    key_list.append(k)

key_list_sorted_asc = sorted(key_list)

value_list = []
for v in ejer_12.values():
    value_list.append(v)

value_list_sorted_desc = sorted(value_list, reverse=True)

print("Lista keys ascendente: ", key_list_sorted_asc)
print("Lista valores descendente: ", value_list_sorted_desc)

ejer_12[1] = 67
print("Clave/valor añadida: ", ejer_12)

print("Valor de la clave 2 del diccionario: ", ejer_12[2])

print("Iterando diccionario")
for k, v in ejer_12.items():
    print("Clave: ", k)
    print("Value: ", v)
````

## Ejercicio 14
Suma todos los valores del dicionario
> ```ejer_14 = {1: 11, 2: 22, 3: 33, 4: 44, 5: 55}```

````python
ejer_14 = {1: 11, 2: 22, 3: 33, 4: 44, 5: 55}
suma = 0
for v in ejer_14.values():
    suma += v

print("Suma total de valores del dict: ", suma)
````
