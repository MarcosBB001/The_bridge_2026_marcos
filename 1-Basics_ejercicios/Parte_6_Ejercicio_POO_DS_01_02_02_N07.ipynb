## Ejercicio 1
Somos una empresa que tiene varias tiendas de electrodomesticos. Necesitamos un programa para manejar las tiendas, ver las ventas que han tenido, cuántos empleados hay, etc... Para ello vamos a modelizar la tienda en una clase, que tendrá las siguientes características:
* Nombre clase: "Tienda"
* Atributos comunes:
    * Tipo: "Electrodomésticos"
    * Abierta: True
* Atributos propios de cada tienda:
    * Nombre - String
    * Dirección - String
    * Número de empleados - int
    * Ventas ultimos 3 meses - Lista de 3 numeros
* Método para calcular las ventas de todos los meses, que devuelva un numero con todas las ventas.
* Método que calcula la media de ventas de los ultimos meses, por empleado
* Método que devuelve en un string el nombre de la tienda, junto con su dirección.
* Método que obtiene las ventas del último mes.
* Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda. Siendo X el parámetro de entrada. Si X es menor de 1000, las ventas de los ultimos 3 meses hubiesen sido de (1.2 x ventas), si es mayor o igual de 1000, las ventas hubiesen sido de (1.5 x venta). El método reescribe el atributo de ventas en los últimos 3 meses, y además devuelve las ventas con el aumento.

Se pide:
1. Implementar la clase "Tienda"

```python
# 1.
class Tienda:
    """Clase Tienda que contiene la informacion de una tienda"""
    def __init__(self, nombre, direccion, n_empleados, ventas_ultimos_3_meses):
        self.tipo = "Electrodomesticos"
        self.abierta = True
        self.nombre = nombre
        self.direccion = direccion
        self.n_empleados = n_empleados  # int
        self.ventas_ultimos_3_meses = ventas_ultimos_3_meses  # Lista de 3 numeros
    
    def calcular_todas_las_ventas(self):
        """Metodo que suma las ventas de todos los meses"""
        ventas_totales = 0
        for venta in self.ventas_ultimos_3_meses:
            ventas_totales += venta
        
        return ventas_totales
    
    def calcular_media_venta_por_empleado(self):
        """Metodo que calcula la media de todas las ventas por empleado"""
        ventas_totales = self.calcular_todas_las_ventas()
        ventas_medias = ventas_totales/self.n_empleados

        return ventas_medias
    
    def get_nombre_y_direccion(self):
        """Metodo que devuelve el nombre de la tienda y la direccion en un solo string"""
        nombre_direccion = "Nombre tienda: " + self.nombre + " --- " + "Dirección: " + self.direccion

        return nombre_direccion
    
    def get_ventas_ultimo_mes(self):
        """Metodo que devuelve las ventas del ultimo mes"""
        return self.ventas_ultimos_3_meses[-1]  # -1 corresponde al último elemento de una lista (por si los últimos 3 meses cambian)
    
    def get_proyeccion_ventas(self, dinero_invertido):
        """
        Metodo que calcula la proyeccion de ventas de cada mes dependiendo de la cantidad de dinero invertida.
            - Si la inversion es menor a 1000, aplica un multiplicador de 1.2 a
            - Si la inversion es mayor a 1000 o igual, aplica un multiplicador de 1.5
        Args:
            dinero_invertido (float): Cantidad de dinero invertido.

        """
        multiplicador = 0
        if dinero_invertido < 1000:
            multiplicador = 1.2
        else:
            multiplicador = 1.5
        
        for i, venta in enumerate(self.ventas_ultimos_3_meses):
            self.ventas_ultimos_3_meses[i] = self.ventas_ultimos_3_meses[i] * multiplicador
        
        return self.ventas_ultimos_3_meses

# 2.
tienda1 = Tienda(nombre="Electrico-domésticos", direccion="Avenida del trueno 7", n_empleados=5, ventas_ultimos_3_meses=[21, 34, 140])
tienda2 = Tienda("Super cocina", "Avenida golosina 5", 10, [100, 200, 100])
tienda3 = Tienda("Memedomésticos", "Calle brainrot 67", 2, [1000, 67, 0])

# 3.
print("##### " + str(3) + " #####")
todas_tiendas = [tienda1, tienda2, tienda3]
for tienda in todas_tiendas:
    print("Nombre: ", tienda.nombre)
    print("Direccion ", tienda.direccion)
    print("Tipo: ", tienda.tipo)
    print("Numero de empleados: ", tienda.n_empleados)
    print("Ventas ultimos 3 meses: ", tienda.ventas_ultimos_3_meses)
    print("Ventas medias por empleado: ", tienda.calcular_media_venta_por_empleado())
    print("Ventas de último mes: ", tienda.get_ventas_ultimo_mes())
    dinero = 1050
    print("Ventas con inversión de 1050 euros: ", tienda.get_proyeccion_ventas(dinero))
    print("¿Esta abierta?: ", tienda.abierta)

print("                ")
# 4. 
print("##### " + str(4) + " #####")
for tienda in todas_tiendas:
    todas_ventas = tienda.calcular_todas_las_ventas()
    print(f"Ventas de la tienda {tienda.nombre}: {todas_ventas}")

print("                ")
# 5. 
print("##### " + str(5) + " #####")
for tienda in todas_tiendas:
    if "Avenida" in tienda.direccion:
        nombre_avenida = tienda.get_nombre_y_direccion()
        print(nombre_avenida)

        
```

## Ejercicio 2
En este ejercicio vamos a implementar una clase *Perro* en Python. La clase tiene las siguientes características:
* Cosas que sabemos seguro que tiene un perro
    * Tiene 4 patas
    * 2 orejas
    * 2 ojos
    * Una velocidad de 0. Por defecto, el perro está parado
* Cuando se inicialice:
    * El perro será de una determinada raza
    * Por defecto tendrá pelo "Marrón", a no ser que se diga lo contrario.
    * Por defecto no tendrá dueño, a no ser que se diga lo contrario.
    
* Dispondrá también de un método llamado andar, que tiene un argumento de entrada (aumento_velocidad). Este valor se le sumará a la velocidad que ya llevaba el perro.
* Necesita otro método (parar), donde pondremos la velocidad a 0.
* Otro método llamado "ladrar", que tendrá un argumento de entrada, y la salida será el siguiente string: "GUAU!" + el argumento de entrada.


Se pide:
* Implementa la clase *Perro*
* Crea un objeto de tipo *Perro*, sin dueño
* Comprueba que están bien todos sus atributos
* Prueba que ande, y comprueba su velocidad
* Páralo
* Documenta la clase *Perro*

```python
# 1.
class Perro:
    """Clase que contiene la informacion de un perro"""
    def __init__(self, raza, pelo="Marron", dueno=None):
        self.raza = raza
        self.pelo = pelo
        self.dueno = dueno
        self.n_patas = 4
        self.n_orejas = 2
        self.n_ojos = 2
        self.velocidad = 0
    
    def andar(self, aumento_velocidad):
        """
        Metodo que suma la velocidad del argumento a la velocidad del perro
        Args:
            aumento_velocidad (int or float): Velocidad a sumar a la velocidad del perro
        """
        self.velocidad += aumento_velocidad
    
    def parar(self):
        """Metodo que pone la velocidad del perro a 0"""
        self.velocidad = 0

    def ladrar(self, frase):
        """
        Metodo que devuelve un string con la forma "GUAU! + <frase>"
        Args:
            frase (str): Frase a poner despues de "GUAU! "
        """
        frase_perro = "GUAU! " + str(frase)

        return frase_perro

# 2.
perro_sin_dueno = Perro("Golden Retriever")
# 3.
print("##### " + str(3) + " #####")
print("Raza: ", perro_sin_dueno.raza)
print("Color pelo: ", perro_sin_dueno.pelo)
print("Dueno: ", perro_sin_dueno.dueno)
print("Numero patas:, ", perro_sin_dueno.n_patas)
print("Numero orejas:, ", perro_sin_dueno.n_orejas)
print("Numero ojos:, ", perro_sin_dueno.n_ojos)
print("Velocidad: ", perro_sin_dueno.velocidad)

print("                 ")
# 4.
print("##### " + str(4) + " #####")
vel = 5
print("Acelerando en ", vel)
perro_sin_dueno.andar(vel)
print("Velocidad actual: ", perro_sin_dueno.velocidad)
vel = 10
print("Acelerando en ", vel)
perro_sin_dueno.andar(vel)
print("Velocidad actual: ", perro_sin_dueno.velocidad)

print("                 ")
# 5.
print("##### " + str(5) + " #####")
print("Parando...")
perro_sin_dueno.parar()
print("Velocidad final: ", perro_sin_dueno.velocidad)



```
